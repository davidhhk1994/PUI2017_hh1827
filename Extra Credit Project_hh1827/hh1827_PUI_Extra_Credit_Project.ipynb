## Ⅲ. Hypothesis Testing: Z Test

#### The null hypothesis will be: The percentage of observations labelled as Unhelathy or worse (PM2.5 Value > 150) during the 2013-2014 period is lower or equal to its counterpart duirng the 2016-2017 period. 


# _$H_0$_ : $\frac{Polluted_{\mathrm{1617}}}{ALL_{\mathrm{1617}}} >= \frac{Polluted_{\mathrm{1314}}}{ALL_{\mathrm{1314}}}$
# _$H_1$_ : $\frac{Polluted_{\mathrm{1617}}}{ALL_{\mathrm{1617}}} < \frac{Polluted_{\mathrm{1314}}}{ALL_{\mathrm{1314}}}$


## I will use a significance level  $\alpha=0.05$

## The statistics is: 
# $z = \frac{(p_0 - p_1)}{SE} $
# $p =\frac{p_0  n_0 + p_1  n_1}{n_0+n_1}$
# $SE = \sqrt{ p  ( 1 - p )  (\frac{1}{n_0} + \frac{1}{n_1}) }$
## and in this notation 
# $p_0 = \frac{Polluted_{\mathrm{1617}}}{ALL_{\mathrm{1617}}}$
### and 
# $p_1 = \frac{Polluted_{\mathrm{1314}}}{ALL_{\mathrm{1314}}}$

```python
count_1617 = len(bj1617)
count_1314 = len(bj1314)

p_1617 = bj1617[bj1617['PM2.5 Value']>150].count()[0] / float(len(bj1617))
p_1314 = bj1314[bj1314["PM2.5 Value"]>150].count()[0] / float(len(bj1314))
print("The total number of observations for 2013-2014 is {}, for 2016-2017 is {}. p0 is {}, p1 is {}.".format(count_1617, count_1314, p_1617, p_1314))
```

```python
p = lambda p0, p1, n0, n1: (p0 * n0 + p1 * n1) / (n0 + n1)
se = lambda p, n0, n1: np.sqrt(p * (1 - p) * (1.0 / n0 + 1.0 / n1)) 
zscore = lambda p0, p1, s : (p0 - p1) / s
```

```python
sp_stdev = se(p(p_1617, p_1314, count_1617, count_1314), count_1617, count_1314)

z = zscore(p_1617, p_1314, sp_stdev)
print ("The z statistics is {}.".format(z))
```

#### Since this calculated Z score is lower than the smallest value in the Z distribution table, it yields a p-value that is smaller than 0.0002, which is significantly smaller than the threshold 0.05. Hence the null hypothesis that portion of heavy pollution observations during 2016-2017 is larger or equal to the counterpart of 2013-2014. The result suggests that 3years after the government's pollution control plan put into practicce, the air quality has been indeed improved.
