### Figure 1: Histogram for original poisson distribution

```python
np.random.seed(999)
L = 900.
Nsamples = 1000
nit = 1000
myDmeans = np.zeros(nit)

for i in range(nit):
    mydistrib = sp.stats.poisson.rvs(mu=L,size = Nsamples)
    myDmeans[i] = mydistrib.mean()
```

```python
pl.hist(myDmeans, bins=30)
pl.xlabel('sample mean', fontsize = 10)
pl.ylabel('N', fontsize = 10)
```
