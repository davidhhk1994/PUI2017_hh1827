```python
import pandas as pd
import numpy as np
import os
from scipy import stats
import matplotlib.pylab as pl
import statsmodels.api as sm
import statsmodels.formula.api as smf
pylab.style.use('ggplot')
%pylab inline
```

### Reading in the NYC building energy peformance disclosure data

```python
# Read data
df_xls = pd.read_excel(r'http://www.nyc.gov/html/gbee/downloads/excel/nyc_benchmarking_disclosure_data_reported_in_2016.xlsx')
# Convert to .csv
df_xls.to_csv(os.getenv("PUIDATA") + '\nyc_benchmarking_disclosure_data_reported_in_2016.csv', encoding='utf-8')
df = pd.read_csv(os.getenv("PUIDATA") + '\nyc_benchmarking_disclosure_data_reported_in_2016.csv')
```

```python
df.head(2)
```

### Data Munging

```python
del df['Unnamed: 0']
# In case the system doesn't recognize the current encoding
cols = df.columns.str.replace('\xc2\xb2', '_2')
df.columns = cols
```

```python
df.head(2)
```

```python
selected = df[['Site EUI (kBtu/ft_2)', 'Total GHG Emissions (Metric Tons CO2e)']]
selected.columns = ["EUI", "Emissions"]
selected.head()
```

```python
selected = selected.dropna(axis=0)
selected.head()
```

```python
selected = selected[~(selected == 0).any(axis=1)]
selected.head()
```

```python
ols = smf.ols(formula=' Emissions~EUI',data=selected).fit()
ols.summary()
```

```python
fig = pl.figure(figsize=(18,8))
pl.scatter(selected.EUI, selected.Emissions, color = "blue", label = "Observations")
pl.plot(selected.EUI, ols.predict(), color = "red", label = 'Prediction line')
pl.title('GHG Emissions VS EUI', fontsize = 22)
pl.xlabel('EUI', fontsize = 15)
pl.ylabel('GHG Emissions', fontsize = 15)
pl.legend(loc='best', fontsize = 18)
pl.show()
```

```python
# Removing outliers in the dimension of EUI
selected = selected[np.abs(selected.EUI-selected.EUI.mean())<=(5*selected.EUI.std())]
```

```python
# Removing outliers in the dimension of GHG Emissions
selected = selected[np.abs(selected.Emissions-selected.Emissions.mean())<=(5*selected.Emissions.std())]
```

```python
curvemodel = smf.ols(formula = 'Emissions~EUI + I(EUI**2)', data = selected).fit()
curvemodel.summary()
```

```python
ols = smf.ols(formula=' Emissions~EUI',data=selected).fit()
ols.summary()
```

```python
fig = pl.figure(figsize=(18,8))
# plotting the observation points
pl.scatter(selected.EUI, selected.Emissions, color = "blue", label = "Observations")
#plotting the linear regression model
pl.plot(selected.EUI, ols.predict(), color = "red", linewidth = 3, label = 'Linrear Regression Model')
#plotting the 2nd degree polynomial model
X = np.linspace(selected['EUI'].min(), selected['EUI'].max(), 1000)
pl.plot(X, curvemodel.predict(exog = dict(EUI = X)), color = "green", linewidth = 3, label = '2nd Degree Polynomial Model')

pl.title('The Correlation Between Green House Gas Emissions and Energy Usage Intensity', fontsize = 26)
pl.xlabel('Site EUI (kBtu/ft_2)', fontsize = 15)
pl.ylabel('Total GHG Emissions (Metric Tons CO2e)', fontsize = 15)
pl.legend(loc='best', fontsize = 18)
pl.show()
```
